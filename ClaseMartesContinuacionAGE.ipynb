```
path = "https://raw.githubusercontent.com/tec03/Datasets/refs/heads/main/datasets/satis.csv"
import pandas as pd
df = pd.read_csv(path, encoding='latin1')

df.head()
```

```
favs = ['Age', 'Gender', 'Ocupation', 'Monthly income', 'Price', 'Quality',
       'Quantity', 'Availability', 'Taste', 'Flavours', 'Customer relation',
       'Customer services', 'Texture', 'Size', 'Overal satisfaction',
       'Consistency', 'Factor', 'Frowned', 'reason', 'Popular', 'My favorite',
       'Kind', 'Preference', 'Recommend']
```

```
ndf = df[favs]
ndf.head()
ndf.isnull().sum()
ndf.dropna(inplace=True, how = 'all')
ndf.dropna(inplace=True, how = 'all', axis = 1)
ndf.drop('reason', axis=1, inplace= True)
ndf.isnull().sum()
```

```
ndf
```

```
def cus_cls(x):
  if 0 <= x <= 20:
    return '<= 20'
  elif 21 <= x <= 30:
    return '21 - 30'
  elif 31 <= x <= 40:
    return '31 - 40'
  elif 41 <= x <= 50:
    return '41 - 50'
  else:
    return '> 50'
```

```
ndf['AgeGp'] = ndf['Age'].apply(cus_cls)
ndf
```

```
counts = pd.pivot_table(ndf, index = 'AgeGp',
               values = 'Gender',
               aggfunc = 'count')
```

```
counts.index
```

```
counts = counts.reindex(['<= 20', '21 - 30', '31 - 40', '41 - 50'])
counts
```

```
type(counts)
counts.columns
counts
```

```
new_name = {'Gender': 'People'}
counts.rename(columns = new_name,
              inplace = True)
counts
```

```
import matplotlib.pyplot as plt
```

```
 explode_list = [0.02, 0.04, 0.06, 0.03]
 color_list = ['gold', 'yellowgreen', 'coral', 'skyblue']

 counts['People'].plot(kind = 'pie',
                      explode = explode_list,
                      colors = color_list,
                      startangle = 90,
                      figsize = [15, 6],
                      autopct = '%1.1f%%',
                      shadow = True,
                      labels = None,
                      pctdistance = 1.1,
                      )

 plt.legend(labels = counts.index,
            loc = 'upper left',
            bbox_to_anchor = (1,0,0.5,1)
            )
 plt.title('Distribution of People by Age Group')
```

```
age_gender = pd.pivot_table(ndf,
               index = ['AgeGp', 'Gender'],
               values = 'Age',
               aggfunc = 'count')
age_gender
```

```
sum = age_gender.iloc[:, 0].sum()
sum
```

```
age_gender['percentage'] = (age_gender['Age'] / sum) * 100
age_gender = age_gender.reset_index()
age_gender
```

```
import plotly.express as px
import seaborn as sns
```

```
age_gender.columns
```

```
plot_order = ['<= 20', '21 - 30', '31 - 40', '41 - 50']
```

```
g = sns.catplot(x = 'AgeGp',
                y = 'percentage',
                data = age_gender,
                kind = 'bar',
                hue = 'Gender',
                order = plot_order,
                palette = 'Set2'
                )

plt.title('Customer Age Groups')
plt.show()
```

```
s = px.sunburst(age_gender,
                path = ['AgeGp', 'Gender'],
                values = 'percentage',
                color = 'percentage',
                color_continuous_scale = 'RdBu',
                title = 'Customer Age Groups'
                )
s.show()
```
